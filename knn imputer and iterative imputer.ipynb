```python
from sklearn.model_selection import train_test_split

from sklearn.impute import KNNImputer,SimpleImputer
```

```python
x = df.drop(['Station_Names'], axis = 1)
```

```python
knn = KNNImputer(n_neighbors=3,weights='distance')

df_knn = knn.fit_transform(x)
```

```python
df_knn = pd.DataFrame(df_knn, columns = ['Year', 'Month', 'Max_Temp','Min_Temp','Rainfall','Relative_Humidity',
           'Wind_Speed','Cloud_Coverage', 'Bright_Sunshine','Station_Number', 'Flood?'])
```

```python
df_knn.to_csv('df_knn_file.csv', header=True, index = False)
```
